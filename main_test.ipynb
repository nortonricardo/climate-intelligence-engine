```python
from pathlib import Path
import pandas as pd

VARIABLES = [
    'temperature',
    'humidity',
    'rainfall',
    'global_radiation',
    'pressure',
]

for var in VARIABLES:
    path = Path(f'data/{var}.parquet')
    if not path.exists():
        print(f'{var}: arquivo não encontrado — rode 1.4_clean_data.py primeiro\n')
        continue
    df = pd.read_parquet(path)
    print(f'=== {var} ({len(df):,} registros) ===')
    print(df.head(10).to_string(index=False))
    print()
```

```python
from pathlib import Path
import pandas as pd

for path in sorted(Path('data').glob('*_neighbors*.parquet')):
    df = pd.read_parquet(path)
    print(f'=== {path.name} ===')
    print(f'Shape  : {df.shape}')
    print(f'Colunas: {df.columns.tolist()}')
    print()
    cols_show = ['code', 'time', 'measurement', 'n01', 'n02', 'd01', 'd02', 'a01', 'a02']
    print(df[cols_show].head(10).to_string(index=False))
    print()
```
